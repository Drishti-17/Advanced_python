**Q. Create a file that contains multiple lines of random strings and file size must be 5 MB.**

---



```python
import random
import string

def generate_random_string(length):
    """Generate a random string of given length"""
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(length))

file_size_limit = 5 * 1024 * 1024  # 5 MB limit
line_length = 100  # Adjust line length as per your requirement

file_name = 'random_file.txt'

with open(file_name, 'w') as file:
    file_size = 0
    while file_size < file_size_limit:
        random_string = generate_random_string(line_length) + '\n'
        file.write(random_string)
        file_size += len(random_string)

print(f'File "{file_name}" created with a size of {file_size} bytes.')
```

**Q. Create 10 files that contains multiple lines of random strings and file size of each file must be 5 MB.**

```python
def convert_to_uppercase(file_name):
    """Convert the contents of a file to uppercase"""
    with open(file_name, 'r+') as file:
        content = file.read().upper()
        file.seek(0)
        file.write(content)

# File names to convert to uppercase
file_names = ['file1GB.txt', 'file2GB.txt', 'file3GB.txt', 'file4GB.txt', 'file5GB.txt']

for file_name in file_names:
    convert_to_uppercase(file_name)

    # Print the file size
    print(f'{file_name}: {os.path.getsize(file_name)} bytes')
```

**Q. WAP to automatically download 10 videos of “Machine Learning” from “Youtube.com”. [Hint: Find the
package from pypi.org and use it]**


```python
!pip install youtube_dl

import youtube_dl

def download_videos(query, limit):
    """Download videos based on the query and limit"""
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',  # Set video format preference
        'outtmpl': '/content/videos/%(title)s.%(ext)s',  # Output directory and filename template
        'ignoreerrors': True,  # Ignore download errors and continue
        'noplaylist': True,  # Avoid downloading playlists
        'max_downloads': limit  # Set the maximum number of downloads
    }

    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        query = f'{query} video'
        search_results = ydl.extract_info(f'ytsearch{limit}:{query}', download=False)

        for video in search_results['entries']:
            if video is not None:
                video_url = video['webpage_url']
                ydl.download([video_url])

query = 'Machine Learning'
limit = 10

download_videos(query, limit)
```

```python

import cv2
import os

# Path to the directory containing the images
image_dir = '/content/images'

# Function to resize an image to 50% of its original size
def resize_image(image_path, save_path):
    # Read the image
    image = cv2.imread(image_path)

    # Get the original image dimensions
    height, width = image.shape[:2]

    # Calculate the new dimensions
    new_width = int(width * 0.5)
    new_height = int(height * 0.5)

    # Resize the image
    resized_image = cv2.resize(image, (new_width, new_height))

    # Save the resized image
    cv2.imwrite(save_path, resized_image)

# Iterate through each image in the directory
for filename in os.listdir(image_dir):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        image_path = os.path.join(image_dir, filename)
        save_path = os.path.join(image_dir, 'resized_' + filename)
        # Resize the image and save it
        resize_image(image_path, save_path)

        print(f"Resized {filename} to 50%.")

print("Resizing complete.")
```

**Q. Create a random dataset of 100 rows and 30 columns. All the values are defined between [1,200]. Perform
the following operations:**



*   (i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number
rows having missing values
*   (ii) Replace all the NA values with the average of the column value.


*   (iii) Find the Pearson correlation among all the columns and plot heat map. Also select those columns
having correlation <=0.7.
*   (iv) Normalize all the values in the dataset between 0 and 10.

*  (v) Replace all the values in the dataset with 1 if value <=0.5 else with 0.







.





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Create a random dataset of 100 rows and 30 columns
dataset = pd.DataFrame(np.random.randint(1, 201, size=(100, 30)))

# (i) Replace values with NA in the dataset defined between [10, 60]
dataset.iloc[10:61] = np.nan
missing_rows_count = dataset.isnull().sum(axis=1).sum()
print(f"Number of rows with missing values: {missing_rows_count}")

# (ii) Replace NA values with the average of the column value
dataset = dataset.fillna(dataset.mean())

# (iii) Find Pearson correlation among columns and plot heatmap
correlation = dataset.corr()
sns.heatmap(correlation, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Pearson Correlation Heatmap')
plt.show()

# Select columns with correlation <= 0.7
columns_with_low_correlation = correlation.columns[correlation.abs().mean() <= 0.7]
print(f"Columns with correlation <= 0.7: {columns_with_low_correlation}")

# (iv) Normalize all values in the dataset between 0 and 10
normalized_dataset = (dataset - dataset.min()) / (dataset.max() - dataset.min()) * 10

# (v) Replace values with 1 if value <= 0.5 else with 0
binary_dataset = normalized_dataset.applymap(lambda x: 1 if x <= 0.5 else 0)

```

**Q. Create a random dataset of 500 rows and 10 columns.**


>

1.   Columns 1 to 4 are defined between [-10, 10];
2.   Columns 5 to 8 are defined between [10, 20];

1.   Columns 9 to 10 are defined between [-100, 100].


Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
graph using each algorithm.


*   (i) K-Mean clustering


*  (ii) Hierarchical clustering












```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.spatial.distance import pdist, squareform
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Create a random dataset
dataset = pd.DataFrame({
    'Column1': np.random.uniform(-10, 10, size=500),
    'Column2': np.random.uniform(-10, 10, size=500),
    'Column3': np.random.uniform(-10, 10, size=500),
    'Column4': np.random.uniform(-10, 10, size=500),
    'Column5': np.random.uniform(10, 20, size=500),
    'Column6': np.random.uniform(10, 20, size=500),
    'Column7': np.random.uniform(10, 20, size=500),
    'Column8': np.random.uniform(10, 20, size=500),
    'Column9': np.random.uniform(-100, 100, size=500),
    'Column10': np.random.uniform(-100, 100, size=500)
})

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(dataset)

# Determine optimal number of clusters for K-Means using the elbow method
inertias = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(dataset)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), inertias, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve for K-Means Clustering')
plt.show()

# Apply Hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
hierarchical_labels = hierarchical.fit_predict(dataset)

# Calculate distance matrix
distances = pdist(dataset, metric='euclidean')
distance_matrix = squareform(distances)

# Plot the distance matrix heatmap
sns.heatmap(distance_matrix, cmap='coolwarm')
plt.title('Distance Matrix Heatmap')
plt.show()
```

**Q. Create a random dataset of 600 rows and 15 columns. All the values are defined between [-100,100].**


Perform the following operations:

(i) Plot scatter graph between Column 5 and Column 6.


(ii) Plot histogram of each column in single graph.


(iii) Plot the Box plot of each column in single graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Create a random dataset
dataset = np.random.uniform(-100, 100, size=(600, 15))

# (i) Plot scatter graph between Column 5 and Column 6
column5 = dataset[:, 4]  # Index starts from 0
column6 = dataset[:, 5]
plt.scatter(column5, column6)
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.title('Scatter Plot: Column 5 vs Column 6')
plt.show()

# (ii) Plot histogram of each column in a single graph
plt.figure(figsize=(10, 6))
for i in range(dataset.shape[1]):
    plt.hist(dataset[:, i], bins=30, alpha=0.5, label=f'Column {i+1}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Each Column')
plt.legend()
plt.show()

# (iii) Plot the Box plot of each column in a single graph
plt.figure(figsize=(10, 6))
plt.boxplot(dataset, labels=[f'Column {i+1}' for i in range(dataset.shape[1])])
plt.xlabel('Column')
plt.ylabel('Value')
plt.title('Box Plot of Each Column')
plt.show()
```

**Q. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].**


Perform the following operations:


(i) Perform t-Test on each column.


(ii) Perform Wilcoxon Signed Rank Test on each column.


(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```python
import numpy as np
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums

# Set random seed for reproducibility
np.random.seed(42)

# Create a random dataset
dataset = np.random.uniform(5, 10, size=(500, 5))

# (i) Perform t-Test on each column
for i in range(dataset.shape[1]):
    col = dataset[:, i]
    t_stat, p_value = ttest_1samp(col, popmean=7.5)  # Assuming population mean of 7.5
    print(f"Column {i+1}: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

# (ii) Perform Wilcoxon Signed Rank Test on each column
for i in range(dataset.shape[1]):
    col = dataset[:, i]
    w_stat, p_value = wilcoxon(col - 7.5)  # Assuming null hypothesis median difference of 0
    print(f"Column {i+1}: Wilcoxon statistic = {w_stat:.4f}, p-value = {p_value:.4f}")

# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
col3 = dataset[:, 2]
col4 = dataset[:, 3]

# Two Sample t-Test
t_stat, p_value = ttest_ind(col3, col4)
print(f"Two Sample t-Test: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

# Wilcoxon Rank Sum Test
w_stat, p_value = ranksums(col3, col4)
print(f"Wilcoxon Rank Sum Test: Wilcoxon statistic = {w_stat:.4f}, p-value = {p_value:.4f}")
```
